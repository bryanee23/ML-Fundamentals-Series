### Approach

- Set x and x-axis using np.arange

Create y values by multiplying equation with x array:
- Set y = $2x$ + 3

Create an array of additional data points using a normal distribution w/ the size of x:
- y_$noise$ = (some integer) * normal distro

Create data of y from the y array and the random y:
- y_$data$ = y + y_$noise$

Create visual:
- Plot a scatter plot parameters(x, y_$data$, 'first letter of color + o')
- Plot a linear regression line w/ (x, y, '- first letter of color')



<hr>

```python
# set x-axis and populate x array
x = np.arange(-5.0, 5.0, 0.1)

# set y array using the linear regression equation * x array
y = 2*(x) + 3

# create random data points
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise

# create scatter plot
plt.plot(x, ydata,  'bo')

# create linear regession
plt.plot(x,y, 'r') 

# overlay both graphs
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

### Cubic Graph Visual

- Same process as above
- Equation is: $ y = (a x^3 + b x^2 + c x + d)$

<hr>

```python
# set x axis w/ an array size of 100 
x = np.arange(-5, 5, .1)

# set y to the equation
a_coef = 1
b_coef = 3
c_coef = 4
d_coef = 4
y = (a_coef * (x**3) + b_coef * (x**2) + (c_coef * x) + d_coef)

# set random y variables

y_noise = random.randint(40,50) * np.random.normal(size=x.size)

# create dataset
y_data = y + y_noise

# create scatterplot
plt.plot(x, y_data, 'bo')

# create best fit line
plt.plot(x, y, '-r')

# overlay both plots
plt.xlabel('Dependent Variables')
plt.ylabel('Independent Variables')
plt.show()
```

### Sigmoidal/Logistic Graph Visual

- Equation is: $ y = a + \frac{b}{1+ c^{(x-d)}}$

Create Visual
- plot(x, y)

From the Numpy docs:

- numpy.power(x1, x2)

    --Return evenly spaced values within a given interval.

- parameters:

    --x1:array like
The bases

    --x2:array like
The exponent

<hr>

```python
x = np.arange(-5, 5, .1)

a_coef = 5
b_coef = 3
c_coef = 3
d_coef = 3

numerator = a_coef
denominator = 1 + np.power(c_coef, (x-d_coef))
y = a_coef + (numerator/denominator) 

plt.plot(x,y)
plt.xlabel('Dependent Variable')
plt.ylabel('Independent Variable')
plt.show()

# if a_coef is negative, graph points upwards
numerator = -1* a_coef
denominator = 1 + np.power(c_coef, (x-d_coef))
y = a_coef + (numerator/denominator) 

plt.plot(x,y)
plt.xlabel('Dependent Variable')
plt.ylabel('Independent Variable')
plt.show()
```

### Approach cont.


Choose a model which may fit into the data set:
- A Logistic graph is the best fit

- Graph can either point upwards or downwards depending on the negative sign

- Equation: $ y = a + \frac{b}{1+ c^{(x-d)}}$

<br>
Reformating the original formula to the following:

$ \hat{Y} = \frac1{1+e^{\beta_1(X-\beta_2)}}$

$\beta_1$: Controls the curve's steepness,

$\beta_2$: Slides the curve on the x-axis.


<br>

- Define a function to be able to create a dynamic plot

<hr>

```python
# def sigmoid(orientation, x_values, beta_1, beta_2):
#     y = 1 / (1 + np.exp(orientation*beta_1*(x_values-beta_2)))
#     return y

def sigmoid(x, Beta_1, Beta_2):
     y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
     return y
    
beta_1 = 0.1
beta_2 = 1990.0

y_predict = sigmoid(x, beta_1, beta_2)

plt.plot(x, y_predict*15000000000000, 'b')
plt.plot(x, y, 'co')

plt.xlabel('Dependent Variable')
plt.ylabel('Independent Variable')
plt.show()
y_predict 
```

### Approach cont.

- import curve_fit from scipy.optimize

From the SciPy docs:

- scipy.optimize.curve_fit(f, xdata, ydata)

    --Use non-linear least squares to fit a function, f, to data.

- parameters:

    --f:function
The model function, f(x, …). It must take the independent variable as the first argument and the parameters to fit as separate remaining argument

    --xdata:array_like or object
The independent variable where the data is measured. Should usually be an M-length sequence or an (k,M)-shaped array for functions with k predictors, but can actually be any object.

    --ydata:array_like
The dependent data, a length M array - nominally f(xdata, ...)

- Returns

    --popt:array
Optimal values for the parameters so that the sum of the squared residuals of f(xdata, *popt) - ydata is minimized.

    --pcov:2-D array
The estimated covariance of popt. The diagonals provide the variance of the parameter estimate. To compute one standard deviation errors on the parameters use perr = np.sqrt(np.diag(pcov))

<hr>

<hr>

### Approach cont.

fitting the curve determines the $\beta$'s needed for the equation

$ \hat{Y} = \frac1{1+e^{\beta_1(X-\beta_2)}}$

$\beta_1$: Controls the curve's steepness,

$\beta_2$: Slides the curve on the x-axis.

<hr>

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, x_data, y_data)
#print the final parameters
print(" beta_1 = %f, beta_2 = %f" % (popt[0], popt[1]))
popt
```


<hr>

### Approach cont.

- with $\beta$'s known

- set the x axis by creating an array using np.linspace to return evenly spaced numbers over a specified interval

- normalize x

- determine y values using sigmoid function


From the docs:

- np.linspace(start, stop, num=50)


- Parameters:

    --start:array_like
The starting value of the sequence.

    --stop:array_like
The end value of the sequence, unless endpoint is set to False. In that case, the sequence consists of all but the last of num + 1 evenly spaced samples, so that stop is excluded. Note that the step size changes when endpoint is False.

    --num:int, optional
Number of samples to generate. Default is 50. Must be non-negative.

- Returns:

    --samples:ndarray
There are num equally spaced samples in the closed interval start, stop or the half-open interval start, stop


<hr>

```python
x_fit = np.linspace(1960, 2015, 60)
x_fit = x_fit/max(x_fit)
y_fit = sigmoid(x_fit, *popt)
```

### Evaluation

Split the dataset into training/testing:
- split = np.random.rand[len(df)] <.80
- train = df[split]
- test = df[~split]

Normalize the data
- test_x = test_x/max(test_x)
- test_y = test_y/max(test_y)

Fit test data onto curve:
-curve_fit(func, test_x(norm), test_y(norm), )

Apply the sigmoid equation onto the testing set:
- $ \hat{Y} = \frac1{1+e^{\beta_1(X-\beta_2)}}$
- use the sigmoid function 

Calculate Errors:
- MSE Mean Squared Error = np.mean((y_predict - y)**2)
- MAE Mean Absolute Error = np.mean(np.absolute(y_predict - y))
- r2_score using sklearn.metrics, r2_score



<hr>

```python
from sklearn.metrics import r2_score

split = np.random.rand(len(df)) <.8

train = df[split]
test = df[~split]

x_test = test['Year'].values
x_test = x_test/max(x_test)
y_test = test['Value'].values
y_test = y_test/max(y_test)

popt, pcov = curve_fit(sigmoid, x_test, y_test) 

y_predict = sigmoid(x_test, *popt)

MSE = np.mean((y_predict - y_test)**2)
MAE = np.mean(np.absolute(y_predict - y_test))
R2_score = r2_score(y_test, y_predict)

print('MSE      = %.3f' % MSE)
print('MAE      = %.3f' % MAE)
print('r2_score = %.3f' % R2_score)
```
